# Importing Neccesary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing The Dataset From CSV File

```python
college_dataset=pd.read_csv("College_data.csv")
college_dataset.head(10)
```

# Checking Size Of The Datset

```python
college_dataset.shape
```

```python
college_dataset.info()
```

```python
college_dataset.dtypes
```

## We can See that All the columns have object data type but Some can have int ,float datatype like UG_fee,rating etc.

```python
college_dataset.isnull().sum()
```

## We can See this Dataset has No Null Values as Null But Missing values are There

```python
mask = college_dataset['UG_fee'].str.contains('--', na=False)
college_dataset[mask]
```

## We can see Many Rows And Columns Contains Null Values in Form of These'--' lets replace them with nan values

```python
import numpy as np
college_dataset=college_dataset.replace('--',np.nan)
college_dataset
```

##### Now Lets See Null Values

```python
college_dataset.isnull().sum()
```

### Now We can replace values with some other values

```python
College_Dataset_NEW=college_dataset.ffill()
College_Dataset_NEW
```

## For Simplicity Stake We Use Forward Fill To replace Null Values
## But Idealy WE Should Use Averge Value For Each Column 

```python
College_Dataset_NEW.isnull().sum()
```

## Now WE Have No Null Values so lets Ckeck for duplicates also

```python
College_Dataset_NEW.duplicated().sum() 
```

#### NO Duplicate Values

### We Replaced The , Values in UG and PG Fees with ' ' and then Convert Them into INT

```python
College_Dataset_NEW['UG_fee'] = College_Dataset_NEW['UG_fee'].str.replace(',', '').astype(int)
College_Dataset_NEW['PG_fee'] = College_Dataset_NEW['PG_fee'].str.replace(',', '').astype(int)
```

### Now Lets Change The Data Types Of some Columns to Numeric

```python
College_Dataset_NEW_INT=College_Dataset_NEW.astype({'Rating': 'float', 'Academic': 'float', 'Faculty': 'float','Infrastructure': 'float', 
                                                    'Placement': 'float','Social_Life': 'float','Accommodation': 'float'})
```

```python
College_Dataset_NEW_INT.dtypes
```

## Now We Finaly Have The Correct Data Types of All The Columns

```python
College_Dataset_NEW_INT
```

### Let's Find Names Of Colleges with Most UG_Fees,PG_fees and Averge Fees

```python
sorted_UG_Fee = College_Dataset_NEW_INT.sort_values(by='UG_fee', ascending=False)
First_10_ROWS_sorted_UG_Fee=sorted_UG_Fee.head(10)
First_10_ROWS_sorted_UG_Fee
```

```python
sorted_PG_Fee = College_Dataset_NEW_INT.sort_values(by='PG_fee', ascending=False)
First_10_ROWS_sorted_PG_Fee=sorted_PG_Fee.head(10)
First_10_ROWS_sorted_PG_Fee
```

```python
College_Dataset_NEW_INT["Averge_Fees"]=(College_Dataset_NEW_INT['UG_fee']+College_Dataset_NEW_INT['PG_fee'])/2
College_Dataset_NEW_INT.head(10)
```

```python
sorted_Average_Fee = College_Dataset_NEW_INT.sort_values(by='Averge_Fees', ascending=False)
First_10_ROWS_sorted_Average_Fee=sorted_Average_Fee.head(10)
First_10_ROWS_sorted_Average_Fee
```

### Now Let's Plot graphs for Each one of Them

```python

plt.xlabel('Colleges_Names')
plt.ylabel('UnderGraduate_Fees')
plt.title('TOP 10 Colleges With THE Most UG Fees')
plt.xticks(rotation=90)  # Rotates labels to 45 degrees
plt.bar(First_10_ROWS_sorted_UG_Fee.College_Name,First_10_ROWS_sorted_UG_Fee.UG_fee, color='skyblue')
plt.show()
```

```python
plt.xlabel('Colleges_Names')
plt.ylabel('PostGraduate_Fees')
plt.title('TOP 10 Colleges With THE Most PG Fees')
plt.xticks(rotation=90)  # Rotates labels to 45 degrees
plt.bar(First_10_ROWS_sorted_PG_Fee.College_Name,First_10_ROWS_sorted_PG_Fee.PG_fee, color='Red')
plt.show()
```

```python
plt.xlabel('Colleges_Names')
plt.ylabel('Average_Fees')
plt.title('TOP 10 Colleges With THE Most Average Fees')
plt.xticks(rotation=90)  # Rotates labels to 45 degrees
plt.bar(First_10_ROWS_sorted_Average_Fee.College_Name,First_10_ROWS_sorted_Average_Fee.Averge_Fees, color='green')
plt.show()
```

### Lets Find The Most popular Stream In each State by Groupby Function

```python
stream_counts = College_Dataset_NEW_INT.groupby(['State', 'Stream']).size().reset_index(name='Counts')
most_popular_stream = stream_counts.sort_values(['State', 'Counts'], ascending=[True, False]).drop_duplicates('State')
print(most_popular_stream)
```

### Lets Plot a BarGraph With This Data  

```python
plt.figure(figsize=(14, 7))
barplot = sns.barplot(data=most_popular_stream, x='State', y='Counts', hue='Stream', dodge=False)
plt.xlabel('State')
plt.ylabel('Counts')
plt.title('Most Popular Stream by State')
plt.xticks(rotation=90)
plt.show()
```

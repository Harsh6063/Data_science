```python
#importing Cars_Prices Dataset
Cars_Price=pd.read_csv("carprices.csv")
Cars_Price.rename(columns={'Car Model':'Car_Model'},inplace=True)
Cars_Price.rename(columns={'Sell Price($)':'Sell_Price'},inplace=True)
Cars_Price.rename(columns={'Age(yrs)':'Age'},inplace=True)
Cars_Price
```

```python
#Merging The Two Dataset and Droping the Former Categorical Value
New_cars=pd.concat([Cars_Price,Cars],axis='columns')
New_cars_Final=New_cars.drop(["Car_Model","Mercedez Benz C class"],axis='columns')
New_cars_Final
```

```python
#Predicting Model
Cars_Model=linear_model.LinearRegression()
Cars_Model.fit(New_cars_Final[['Mileage','Age','Audi A5','BMW X5']],New_cars_Final.Sell_Price)
```

```python
Cars_Model.coef_ #Finds The value of m 
```

```python
Cars_Model.intercept_ #Finds The value of b
```

```python
#Finding The Reliability of The Model
Cars_Model.score(New_cars_Final[['Mileage','Age','Audi A5','BMW X5']],New_cars_Final.Sell_Price)
#Its Giving 94% Accurate
```

```python
#Predicting Sell Price
Cars_Model.predict([[40000,4,0,1]])
```

```python
#Predicting The Output
Survive_Predict.predict([[1,21,55,1]]) # 1 Here Means True
```

```python
# Measuring The Accuracy
Survive_Predict.score(inputs,output)
#Its 97.4% Accurate
```

```python
#Importing Digits Datset
from sklearn.datasets import load_digits
DIGITS_DATASET0=load_digits()
dir(DIGITS_DATASET0)
```

```python
#Doing PreProcessing in the Dataset
import pandas as pd
DIGITS_INPUTS=pd.DataFrame(DIGITS_DATASET0.data,columns=DIGITS_DATASET0.feature_names)
DIGITS_INPUTS.head()
```

```python
#Making outputs
DIGITS_OUTPUTS=DIGITS_DATASET0.target
DIGITS_OUTPUTS
```

```python
#Splitting The Dataset into Train And Test
from sklearn.model_selection import train_test_split
Digitsx_train,Digitsx_test,Digitsy_train,Digitsy_test= train_test_split(DIGITS_INPUTS,DIGITS_OUTPUTS,train_size=0.3,random_state=10)
```

```python
#Importing K Nearsest Neighbors Algorithm
from sklearn.neighbors import KNeighborsClassifier
K_predict=KNeighborsClassifier(n_neighbors=3)
K_predict.fit(Digitsx_train,Digitsy_train)
```

```python
#Predicting Scores
K_predict.score(Digitsx_train,Digitsy_train)
```

```python
#Predicting Values
K_predict.predict(Digitsx_test)
```

```python
#Checking Values
Digitsy_test
```

```python
#MAking Confusion_Matrix For Checking The Algorithm 
from sklearn.metrics import confusion_matrix
y_predicteddd=K_predict.predict(Digitsx_test)
cfm=confusion_matrix(Digitsy_test,y_predicteddd)
cfm
```

```python
#Plotting Confusion_Matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(cfm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.show()
```
